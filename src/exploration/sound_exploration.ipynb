```python
import numpy as np
import json
import ipywidgets as widgets
from IPython.display import display, Audio
import random
```

```python

data_path = '../data/16k_data/'  # or './48k'
arrays = np.load(f'{data_path}/training_25112024.npz')
meta = json.load(open(f'{data_path}/meta.json', 'r'))


```

```python

class_indices = {label: [] for label in meta['labels']}
```

```python
for i, label in enumerate(arrays['y']):
    label_name = list(meta['labels'].keys())[label]
    class_indices[label_name].append(i)

```

```python
def play_example(class_name):
    idx = random.choice(class_indices[class_name])
    audio = arrays['X'][idx]
    # Normalize audio
    audio = audio / np.max(np.abs(audio))
    display(Audio(audio, rate=meta['sample_rate']))
    print(f"Playing {class_name}, index: {idx}")
```

```python

for class_name in meta['labels']:
    print(f"\n=== {class_name} ===")
    play_example(class_name)
```
